```python
def square(n):
    return n * n
print(square(4))   # Output: 16
```

```python
def c_to_f(celsius):
    return celsius * 9/5 + 32
print(c_to_f(0))    # Output: 32
print(c_to_f(25))   # Output: 77.0
```

```python
def largest(a, b, c):
    return max(a, b, c)
print(largest(4, 9, 2))   # Output: 9
```

```python
def is_strong(password):
    has_digit = any(ch.isdigit() for ch in password)
    has_upper = any(ch.isupper() for ch in password)
    return len(password) >= 8 and has_digit and has_upper
print(is_strong("Hello123"))   # True
print(is_strong("weakpass"))   # False
```

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    return sum(ch in vowels for ch in text)
print(count_vowels("Python is fun"))   # Output: 4
```
