```python
numbers = input("Enter numbers separated by spaces: ")

# Convert to list of integers
num_list = [int(num) for num in numbers.split()]

# Create a new list without duplicates
unique_list = []
for n in num_list:
    if n not in unique_list:
        unique_list.append(n)

print("Original list:", num_list)
print("List without duplicates:", unique_list)
```

```python
number = input("Enter number by separated by spaces:")
num_list = [int(num) for num in number.split()]

value = int(input("Enter the value ti finde:"))
if value in num_list:
    index_position = num_list.index(value)
    print(f"The value {value} is at index {index_position}.")
else:
    print(f"The value {value} was not found in the list.")
```
