# 1️⃣ Longest Unique Substring

**Given a string, find the longest substring that contains no repeating characters.
Return both the substring and its length.**

```python
def longest_unique_substring(s):
    char_index = {}    # stores last seen index of each character
    left = 0
    max_len = 0
    start = 0          # to record the starting index of the best substring

    for right, char in enumerate(s):
        # If character is repeated AND inside current window
        if char in char_index and char_index[char] >= left:
            left = char_index[char] + 1   # move window

        char_index[char] = right          # update last seen

        # Check if we found a longer substring
        if right - left + 1 > max_len:
            max_len = right - left + 1
            start = left

    return s[start:start + max_len], max_len


# Example
string = "abcabcbb"
substring, length = longest_unique_substring(string)
print("Longest unique substring:", substring)
print("Length:", length)
```

# 2️⃣ Matrix Spiral Traversal

**Given a 2D matrix, print its elements in spiral order.**

```python
def spiral_order(matrix):
    if not matrix:
        return []

    res = []
    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1

    while top <= bottom and left <= right:

        # 1. Left → Right
        for col in range(left, right + 1):
            res.append(matrix[top][col])
        top += 1

        # 2. Top → Bottom
        for row in range(top, bottom + 1):
            res.append(matrix[row][right])
        right -= 1

        # 3. Right → Left  (only if remaining row exists)
        if top <= bottom:
            for col in range(right, left - 1, -1):
                res.append(matrix[bottom][col])
            bottom -= 1

        # 4. Bottom → Top (only if remaining column exists)
        if left <= right:
            for row in range(bottom, top - 1, -1):
                res.append(matrix[row][left])
            left += 1

    return res


# Example
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(spiral_order(matrix))
```

# 3️⃣ Anagram Grouping

**Given a list of words, group them into sets of anagrams.
Example:**

```python
from collections import defaultdict

def group_anagrams(words):
    groups = defaultdict(list)

    for word in words:
        key = ''.join(sorted(word))  # sorted letters used as signature
        groups[key].append(word)

    return list(groups.values())


# Example
words = ["eat", "tea", "ate", "bat", "tab", "tan", "nat"]
print(group_anagrams(words))
```
