```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Example
print(count_vowels("Hello Python"))
```

```python
def is_anagram(str1, str2):
    return sorted(str1) == sorted(str2)

# Example
print(is_anagram("listen", "silent"))
```
