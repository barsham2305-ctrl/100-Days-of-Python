# 1. Find the Largest of Three Numbers

**Take three numbers as input and print the largest.**

```python
a =float(input("Enter first number:"))
b =float(input("Enter second number:"))
c =float(input("Enter third number:"))
if a>=b and a>=c:
    print("The largest number is:", a)
elif b>=a and b>=c:
    print("The largest number is:",b)
else:
    print("The largest number is:",c)
    
```

# 2Check Voting Eligibility

_“Eligible to vote” if age ≥ 18_

# 3Positive, Negative, or Zero

**Input a number and print whether it's positive, negative, or zero.**

```python
num = float(input("Enter a number"))
if num>0:
    print("positive")
elif num<0:
    print("negative")
else:
    print("zero")
```

# 4. Leap Year Checker

**Ask the user for a year and check if it is a leap year.**

```python
year = int(input("enter a yera"))
if( year% 400==0):
    print("Leap year")
elif (year% 100==0):
    print("Not a leap year")
elif (year% 4==0):
    print("Leap year")
else:
    print("Not a leap year")
```

# 5Temperature Converter

**Ask the user for temperature in Celsius and convert to Fahrenheit.**

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit:", fahrenheit)
```
